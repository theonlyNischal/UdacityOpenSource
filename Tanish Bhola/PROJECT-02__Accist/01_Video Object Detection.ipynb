```python
from imageai.Detection import VideoObjectDetection
import os
```

```python
video_detector = VideoObjectDetection()
video_detector.setModelTypeAsYOLOv3()
video_detector.setModelPath("./assets/yolo.h5")
```

```python
video_detector.loadModel()
```

```python
custom_objects = video_detector.CustomObjects(car=True)
```

```python
detectedObjects = []
def forFrame(frame_number, output_array, output_count):
    print("FOR FRAME " , frame_number)
    print("Output for each object : ", output_array)
    print("Output count for unique objects : ", output_count)
    print("------------END OF A FRAME --------------")
    detectedObjects.append(output_array)
```

```python
video_detector.detectCustomObjectsFromVideo(
    custom_objects=custom_objects,
    input_file_path="./assets/vid.mp4",
    output_file_path="./assets/ObjectDetectedVideo",
    frames_per_second=10,
    per_frame_function=forFrame,
    minimum_percentage_probability=30
)
```

```python
len(detectedObjects)
```

## Saving Object list

```python
import pickle
```

```python
with open('./assets/detectedObjects.txt','wb') as fp:
    pickle.dump(detectedObjects, fp)
```
